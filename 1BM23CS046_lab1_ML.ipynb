```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler, StandardScaler
adult_df = pd.read_csv("/content/adult.csv.zip")
print("Dataset Info:")
adult_df.info()
print("\nFirst 5 rows:")
print(adult_df.head())
adult_df.replace('?', np.nan, inplace=True)
for col in adult_df.select_dtypes(include='object').columns:
    adult_df[col].fillna(adult_df[col].mode()[0], inplace=True)
le = LabelEncoder()
for col in adult_df.select_dtypes(include='object').columns:
    adult_df[col] = le.fit_transform(adult_df[col])
numeric_cols = adult_df.select_dtypes(include=np.number)
Q1 = numeric_cols.quantile(0.25)
Q3 = numeric_cols.quantile(0.75)
IQR = Q3 - Q1
numeric_cols = numeric_cols[
    ~((numeric_cols < (Q1 - 1.5 * IQR)) | (numeric_cols > (Q3 + 1.5 * IQR))).any(axis=1)
]
minmax_scaler = MinMaxScaler()
adult_minmax = minmax_scaler.fit_transform(numeric_cols)
standard_scaler = StandardScaler()
adult_standardized = standard_scaler.fit_transform(numeric_cols)

```
